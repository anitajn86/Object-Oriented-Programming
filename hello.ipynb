SCHOOL SYSTEM MODEL

```python
class Student:
    def __init__(self,name,age,grade):
        self.name=name
        self.age=age
        self.grade=grade
    
    def get_grade(self):
        return self.grade
    #this is a method that returns a student's grade

class Course:
    def __init__(self,name,max_students):
        self.name=name
        self.max_students=max_students
        self.students=[]

    def add_student(self,student):
        if len(self.students)<self.max_students:
            self.students.append(student)
            return True
        return False
    def get_average_grade(self):  #average grade of all students enrolled in course
        value=0
        for student in self.students:
            value+=student.get_grade()
        return value/len(self.students)

s1=Student("Alexa",18,99)
s2=Student("Wellerman",20,100)
s3=Student("Johnny Depp",21,100)
print(s1.name,s1.age,s1.grade)
print(s2.name,s2.age,s2.grade)
print(s3.name,s3.age,s3.grade)

course=Course("Science",2)
course.add_student(s1)
course.add_student(s3)
print(course.students[0].name)
print(course.get_average_grade())

        
```

INHERITANCE

```python
class Pet:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def show(self):
        print(f"I am {self.name} and I am {self.age} years old")

    def speak(self):
        print("I don't know what to say")

#Pet is the general class and is being parented by the class Cat which inherits methods from it
class Cat(Pet):
    def speak(self):
        print("Meow")

class Dog(Pet):
    def speak(self):
        print('Bark')

class Fish(Pet):
    pass

#to create an instance for the class Pet
p=Pet('Alexa',20)
p.show()
#now to show inheritance
c=Cat("Jinx",16)
c.show()
c.speak()

f=Fish("Bubbles",10)
f.show()
f.speak()
```

```python
class Pet:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def show(self):
        print(f"I am {self.name} and I am {self.age} years old")

    def speak(self):
        print("I don't know what to say")

#Pet is the general class and is being parented by the class Cat which inherits methods from it
class Cat(Pet):
    def __init__(self,name,age,color):
        super().__init__(name,age)  #super is referencing the parent class. this is to call all the methods and functions from that class to the current class.
        self.color=color

    def speak(self):
        print("Meow")

    def show(self):
        print(f"I am {self.name} and I am {self.age} years old and I am {self.color}")

class Dog(Pet):
    def speak(self):
        print('Bark')

c=Cat("Alexa",19,"Pink")
c.speak()
c.show()
```
