```python
#positional parameter passing
def introduction(firstname, lastname):
    print("Hello, my name is ",firstname, lastname)

introduction("Luke","Skywalker")
introduction("Darth","Vader")
introduction("Kent","Clark")
```

```python
#keyword argument passing
def introduction(firstname, lastname):
    print("Hello, my name is ",firstname, lastname)
introduction(firstname="James" ,lastname="Bond")
introduction(lastname="Kingsley",firstname="Ben")
```

```python
#Mixed
def adding(a,b,c):
    print(a,"+",b,"+",c,"=",a+b+c)
#positional
adding(1,2,3)
#keyword
adding(c=2,a=3,b=1)
#mixed
#put the positional arguments first
adding(2,b=3,c=1)
```
