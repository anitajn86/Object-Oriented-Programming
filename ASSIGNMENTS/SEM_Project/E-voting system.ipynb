```python
#Voter class

class Voter:
    def __init__(self, voter_id, name, age, gender, location):
        self.voter_id = voter_id
        self.name = name
        self.age = age
        self.gender = gender
        self.location = location
        self.has_voted = False
        
        
    def register_voter(self):
        print(f"The Voter {self.name}, ID {self.voter_id} is now registered to vote")
        return True
    
    def vote(self):
        if not self.has_voted:
            self.has_voted = True
            print(f"\nVoter {self.name}, ID {self.voter_id} has successfully voted.")
            return True
        else:
            print(f"\nVoter {self.name}, ID {self.voter_id} has already voted.")
            return False
```

```python
#Candidate class

class Candidate:
    def __init__(self, candidate_id, name, position):
        self.candidate_id = candidate_id
        self.name = name
        self.position = position
        self.vote_count = 0

    def receive_vote(self):
        self.vote_count += 1
```
