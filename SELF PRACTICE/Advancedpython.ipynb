```python
#dunders __double underscores / magic methods
#these define how objects behave with respect to specific python operations
class Person:
    def __init__(self,name,age): #constructor method, called when an object is instantiated
        self.name=name
        self.age=age

    def __del__(self):
        print("Object is being deconstructed")

p=Person("Mike",30) #creating an instance by calling the class Person triggers the init method

        
#__str__(self)defines string representation of an object when print() is called or when object is converted
#  to a string using str()
class Person:
    def _str_(self):
        return f'{self.name}, {self.age}'

#__add__(self,other) defines behaviour for the + operator
class Number:
    def _init_(self, value):
        self.value = value

    def _add_(self, other):
        return self.value + other.value

#__len__(self) called when len()is used on an object
class MyList:
    def _init_(self, data):
        self.data = data

    def _len_(self):
        return len(self.data)

#__eq__(self,other) defines behaviour for the equality operator==
class Person:
    def _init_(self, name, age):
        self.name = name
        self.age = age

    def _eq_(self, other):
        return self.name == other.name and self.age == other.age
    
# __repr__(self) tells python how to represent the vector when we want to represent it.
class Person:
    def __repr__(self):
        return f'Person(name={self.name}, age={self.age})'


```

```python
#dunders __double underscores / magic methods
class vector:
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def __add__(self,other):
        return vector(self.x+other.x, self.y+other.y)
    
    def __repr__(self):
        return f"X:{self.x}, Y: {self.y}"
    
    def __len__(self):
        return 10
    
    def __call__(self):
        print("Hello, I was called")


    
v1=vector(10,20)
v2=vector(50,60)
v3=v1+v2
#print(v3)
#print(len(v3))

#to use the call method
v3() 

#to use the repr method
#print(v3.x)
#print(v3.y)   
#     
```

```python
#Decorators
#add a certain functionality to a function
def my_decorator(function):
    #we give the decorator function another function to which it will add functionality
    def wrapper():  #has/adds some functionality and then it calls the function
        print('I am decorating your function!')
        function()

    return wrapper


```
